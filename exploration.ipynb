```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
def recommander_offres(profil,ville):
    if ville != 'peu importe':
        offres_filtrées = offres[offres['ville']==ville].reset_index(drop=True)
    else:
        offres_filtrées = offres.reset_index(drop=True)
    if len(offres_filtrées)==0:
        return "aucune offre trouvee dans cette ville"
    matrice_offres_filtrées = vectorizer.transform(offres_filtrées['texte_complet'])
    vector_profil = vectorizer.transform([profil.lower()])
    scores= cosine_similarity(vector_profil, matrice_offres_filtrées)
    top_indices = np.argsort(scores[0])[::-1][:3]
    return offres_filtrées.iloc[top_indices][['titre', 'entreprise', 'domaine', 'ville']]
```

```python
synonymes = {
    # Développement / Dev Web
    'dev': 'développement',
    'devweb': 'développement web',
    'front': 'frontend',
    'back': 'backend',
    'bdd': 'base de données',
    'bd': 'base de données',

    # Data / IA
    'ia': 'intelligence artificielle',
    'ai': 'intelligence artificielle',
    'ml': 'machine learning',
    'dl': 'deep learning',
    'data science': 'science des données',
    'analyste de données': 'data analyst',

    # Langages/Frameworks (variantes de casse ou d'écriture)
    'js': 'javascript',
    'ts': 'typescript',
    'py': 'python',

    # DevOps / Cloud
    'ci/cd': 'intégration continue',
    'devops': 'développement opérations',

    # Cybersécurité
    'cyber': 'cybersécurité',
    'sécu': 'sécurité',
    'pentest': 'test de pénétration',

    # Mobile
    'mobile dev': 'développement mobile',
    'app': 'application',

    # Termes génériques de stage
    'pfe': 'projet de fin d\'études',
    'stg': 'stage',
}
```

```python
import re

for abbr in synonymes.keys():
    pattern = r'\b' + re.escape(abbr) + r'\b'
    present_offres = offres['texte_complet'].str.contains(pattern, regex=True).sum()
    present_etudiants = etudiants['texte_complet'].str.contains(pattern, regex=True).sum()
    if present_offres > 0 or present_etudiants > 0:
        print(f"'{abbr}' trouvé : {present_offres} offres, {present_etudiants} étudiants")
```

```python
from synonymes import SYNONYMES

def normaliser_synonymes(texte):
    for mot, remplacement in SYNONYMES.items():
        texte = re.sub(r'\b' + re.escape(mot) + r'\b', remplacement, texte)
    return texte
```
