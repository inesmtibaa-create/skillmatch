```python


import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

offres = pd.read_csv('data/offres.csv')
offres = offres.dropna(subset=['description', 'domaine'])

offres['texte'] = (offres['description'] + ' ' + offres['competences']).str.lower()

X_text = offres['texte']
y = offres['domaine']

print(y.value_counts())
```

```python
from modele import charger_offres, entrainer_vectorizer, recommander_offres

offres = charger_offres()
```

```python
vectorizer = entrainer_vectorizer(offres)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(X_text)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Train:", X_train.shape[0], "exemples")
print("Test:", X_test.shape[0], "exemples")
```

```python
from sklearn.naive_bayes import MultinomialNB

modele_nb = MultinomialNB()
modele_nb.fit(X_train, y_train)

predictions_nb = modele_nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, predictions_nb))
print("\nRapport détaillé:\n", classification_report(y_test, predictions_nb))
```

```python
from sklearn.linear_model import LogisticRegression

modele_lr = LogisticRegression(max_iter=1000)
modele_lr.fit(X_train, y_train)

predictions_lr = modele_lr.predict(X_test)

print("Accuracy Logistic Regression:", accuracy_score(y_test, predictions_lr))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, predictions_nb, xticks_rotation=45)
plt.title("Matrice de confusion - Naive Bayes")
plt.tight_layout()
plt.show()
```
